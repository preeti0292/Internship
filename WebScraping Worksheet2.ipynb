# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
#Checking review
rev = driver.find_elements(By.CLASS_NAME, '_2wzgFH')
a = rev[0].text.split('\n')
a
```

```python
#Checking the length of only review
len(a)-4
```

```python
#Checking Nxt button tags
driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[12]').click()
```

```python
#Defining a function to automate the process of fetching 100 review details and clicking next button everytime untill 10th page.
def iphone_scrapper():
    #Creating empty list
    rating = []
    summary = []
    review = []
    
    #This will have complete review details 
    reve = driver.find_elements(By.CLASS_NAME, '_2wzgFH')
    
    #below for loop will interate the code again again untill review details are fetched
    for index in range(len(reve)):
        items = reve[index].text.split('\n')
        rating.append(items[0])
        summary.append(items[1])
        review.append(''.join(items[2:len(items)-4]).replace('',''))
    
    try:
        #This try section will click on next button for rest of the pages
        driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[12]').click()
    
    except:
        #This is to click on next button for the first page
        driver.find_element(By.CLASS_NAME, '_1LKTO3').click()
        
    time.sleep(2)
    
    return rating, summary, review
```

```python
#Scraping 100 review details.

time.sleep(2)

Rating = []
Summary = []
Review = []

for index in range(0,10):
    ra, su, re = iphone_scrapper()
    Rating.extend(ra)
    Summary.extend(su)
    Review.extend(re)
    
    print(f'Review Page {index + 1} Scrapped')
```

```python
#Creating Dataframe to showcase all review details in one frame.
ipd = pd.DataFrame({'IPhone Ratings':Rating, 'Review Summary':Summary, 'Reviews':Review})
ipd = ipd.style.hide_index()
ipd
```
